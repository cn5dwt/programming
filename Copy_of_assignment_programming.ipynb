In this question, you're going to write a function that finds the highest possible value for some function, and returns the value of the function and the maximizing value.

1. Write a function that creates a grid from some value $a$ to a second value $b$ with $N$ steps, computes the value of a function $f$ on that grid, and then finds the maximizers of the function. Have it return a dictionary with the maximum value and the maximizers themselves.
2. Use your function to maximize $f(x) = 100 - 2x^2 + 3x$ for values of $a=-1$, $b=1$, and for values of $N=3$, $N=10$, $N=100$, $N=1000$ and $N=5000$.
3. The true maximizer of this function is $.75$ and the true maximum value is $101.125$. Is that what you got in the previous step? Why not? How does the quality of the maximization depend on $N$? How does your computed answer change as $N$ gets larger?

````python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  y = 100 - 2*x**2 + 3*x # defining f(x) for maximizer function
  return y

def maximizer(a, b, N, f, showplot=False):
  x = np.linspace(a, b, N) # create a grid from a-b with N steps
  y = f(x) # compute the value of function f
  if showplot == True: # This is a plot that appears when showplot==True and not otherwise
        plt.scatter(x,y)
        plt.show()
  maximum_value = max(y) # the highest value of f(x) on the plot
  maximum_index = np.where(maximum_value==y) # find the index on when y is the maximum value on the plot
  maximizer = x[maximum_index] # computes the x-value associated with the x-index
  return({'maximum': maximum_value, 'maximizer': maximizer}) # this function returns a dictionary of the maximum and maximizer

a = -1
b = 1

print(maximizer(a, b, 3, f, showplot=True))
print(maximizer(a, b, 10, f, showplot=True))
print(maximizer(a, b, 100, f, showplot=True))
print(maximizer(a, b, 1000, f, showplot=True))
print(maximizer(a, b, 5000, f, showplot=True))

f(0.75) # actual maximum value, the value was not exact 101.125 since it is an estimation, but as N gets larger, the estimation gets more accurate. At N = 5000, the esimation was very accurate.
# Question: why is it "showplot=False" when creating the function maximizer, but in the print functions, we do "showplot=True"? Why not make it "showplot=True" when creating maximizer?
````
